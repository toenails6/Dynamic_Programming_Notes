# Count number of ways to construct string example. 
Write a function `countConstruct(target, wordBank)` that accepts a target string and an array of strings. 

The function should return the number of ways that the `target` can be constructed by concatenating elements of the `wordBank` array. 

You may reuse elements of `wordBank` as many times as needed. 

`countConstruct("abcdef", ["ab", "abc", "cd", "def", "abcd"])` $\rightarrow$ $1$ \
<img src="Images/countConstruct_img1.png" alt="drawing" width="400"/>

```python
def countConstruct(target:str, wordBank:list[str], memo={}) -> int:
    if target in memo:
        return memo[target]
    if target == "":
        return 1

    totalCount = 0

    for word in wordBank:
        if target.startswith(word):
            numWaysForTheRest = countConstruct(
                target[len(word):], 
                wordBank, 
                memo)
            totalCount += numWaysForTheRest

    memo[target] = totalCount
    return totalCount

countConstruct("abcdef", ["ab", "abc", "cd", "def", "abcd"])
```

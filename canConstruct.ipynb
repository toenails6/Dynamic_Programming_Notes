# Whether can construct string example. 
Write a function `canConstruct(target, wordBank)` that accepts a target string and an array of strings. 

The function should return a boolean indicating whether or not the `target` can be constructed by concatenating elements of the `wordBank` array. 

You may reuse elements of `wordBank` as many times as needed. 

`canConstruct("abcdef", ["ab", "abc", "cd", "def", "abcd"])` $\rightarrow$ `True` \
<img src="Images/canConstruct_img2.png" alt="drawing" width="400"/>

Note here that one should not take out strings from the middle of the target string, as this could create new adjacencies, instead keep to the prefix of the string recursively. \
<img src="Images/canConstruct_img2.png" alt="drawing" width="400"/>

```python
def canConstruct(target: str, wordBank: list[str], memo={}) -> bool:
    if target in memo: 
        return memo[target]
    if target == "":
        return True

    for word in wordBank:
        if target.startswith(word):
            suffix = target[len(word):]
            if canConstruct(suffix, wordBank):
                memo[target] = True
                return True
            
    memo[target] = False
    return False

canConstruct("abcdef", ["ab", "abc", "cd", "def", "abcd"], memo={})
```
